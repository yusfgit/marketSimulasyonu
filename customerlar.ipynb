```python
import random
import time
import pandas as pd
import matplotlib.pyplot as plt

sexes = ["Man", "Women"]
simulasyon_saati = 8 * 60  # Dakika cinsinden 08:00
kapanis_saati = 22 * 60  # 22:00
day = 1
visitCount = 0
customers = []
total = 0
class Customer():
    def __init__(self, income, age, sex,spent, checkinTime, checkOutTime):
        self.income = income
        self.age = age
        self.sex = sex
        self.checkinTime = checkinTime
        self.checkOutTime = checkOutTime
        self.spent = spent
def plot_statistics(customers, day):
    # Harcama – Müşteri No grafiği
    harcamalar = [c.spent for c in customers]
    musteri_nolar = list(range(1, len(customers) + 1))

    plt.figure(figsize=(12, 8))
    
    # 1. Harcama - Müşteri No
    plt.subplot(2, 2, 1)
    plt.plot(musteri_nolar, harcamalar, marker='o')
    plt.title(f"Gün {day} - Müşteri Harcamaları")
    plt.xlabel("Müşteri No")
    plt.ylabel("Harcama (TL)")
    
    # 2. Cinsiyete Göre Ortalama Harcama
    erkekler = [c.spent for c in customers if c.sex == "Man"]
    kadinlar = [c.spent for c in customers if c.sex == "Women"]
    
    plt.subplot(2, 2, 2)
    plt.bar(["Erkek", "Kadın"], [sum(erkekler)/len(erkekler) if erkekler else 0,
                                 sum(kadinlar)/len(kadinlar) if kadinlar else 0])
    plt.title("Cinsiyete Göre Ortalama Harcama")
    plt.ylabel("Harcama (TL)")

    # 3. Giriş Saatlerine Göre Müşteri Sayısı
    saatlere_gore = {}
    for c in customers:
        saat = c.checkinTime // 60
        saatlere_gore[saat] = saatlere_gore.get(saat, 0) + 1

    plt.subplot(2, 2, 3)
    plt.bar(saatlere_gore.keys(), saatlere_gore.values())
    plt.title("Saat Bazlı Müşteri Girişi")
    plt.xlabel("Saat")
    plt.ylabel("Müşteri Sayısı")

    # 4. Cinsiyete Göre Ortalama Ziyaret Süresi
    erkek_sure = [c.checkOutTime - c.checkinTime for c in customers if c.sex == "Man"]
    kadin_sure = [c.checkOutTime - c.checkinTime for c in customers if c.sex == "Women"]

    plt.subplot(2, 2, 4)
    plt.bar(["Erkek", "Kadın"], [
        sum(erkek_sure)/len(erkek_sure) if erkek_sure else 0,
        sum(kadin_sure)/len(kadin_sure) if kadin_sure else 0
    ])
    plt.title("Cinsiyete Göre Ortalama Kalma Süresi")
    plt.ylabel("Dakika")

    plt.tight_layout()
    plt.show()
def generate_customer(checkinTime):
    """Rastgele müşteri oluşturur ve giriş saatine göre ayarlar."""
    sex = random.choice(sexes)
    income = random.randint(5000, 25000)  # Rastgele gelir
    age = random.randint(18, 80)  # Rastgele yaş
    shopping_duration = random.randint(10, 120)  # 10-120 dakika alışveriş süresi
    checkOutTime = min(checkinTime + shopping_duration, kapanis_saati)  # Market kapanmadan çıkış
    spent = random.randint(0,income)
    return Customer(income, age, sex,spent, checkinTime, checkOutTime)
while True:
    time.sleep(0.01)  
    simulasyon_saati += 1
      # %30 ihtimalle yeni müşteri oluştur
    if random.random() < 0.3:
        customer = generate_customer(simulasyon_saati)
        customers.append(customer)
        visitCount += 1
        print(f"Yeni müşteri geldi ve Bilgileri:\nCinsiyeti: {customer.sex}\nYaşı:{customer.age}\nGeliri:{customer.income}\nHarcanan: {customer.spent}\nGiriş Saati - Çıkış Saati: {customer.checkinTime // 60}:{customer.checkinTime % 60:02d} - {customer.checkOutTime // 60}:{customer.checkOutTime% 60:02d} \ Gün : {day}\nMüşteri Sırası: {visitCount}")
        total += customer.spent
    # Gün sonu kontrolü
    if simulasyon_saati >= kapanis_saati:
        print(f"GÜN {day} BİTTİ! Toplam {len(customers)} müşteri geldi.\nTOTAL: {total}")
        day += 1
        simulasyon_saati = 8 * 60  # Yeni gün 08:00'de başlasın
        plot_statistics(customers, day)

```
